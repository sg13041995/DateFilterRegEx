### Combining the four steps in a single solution program

```python
import re
import linecache
from dateutil import parser
from datetime import datetime

# initializing empty dict to capture {line_number : date_data}
result_dict = {}
# total line numbers
total_line_numbers_list = [line for line in range(1,501)]
# line numbers where we havent received any data
empty_line_numbers_list = []

# data filter function that returns the filtered values in resultant dictionary and also the line numbers where data was unable to filter for the given regex, for the subsequent filtering processes
def filter_data(regex_pattern, total_line_numbers_list, empty_line_numbers_list, result_dict):
    
    for line in total_line_numbers_list:
        line_data = linecache.getline("./Raw_Medical_Data_INPUT.txt", line)
        matched_data = regex_pattern.findall(line_data)
        
        # getting the line numbers where no data found 
        if matched_data == []:
            empty_line_numbers_list.append(line)
        else:
            # else getting the data in the dictionary
            if type(matched_data[0]) == tuple:
                if matched_data[0][0] == '':
                    result_dict[line] = matched_data[0][1]
                elif matched_data[0][1] == '':
                    result_dict[line] = matched_data[0][0]
            elif type(matched_data[0]) == str:
                result_dict[line] = matched_data[0]
                
    return empty_line_numbers_list, result_dict

# regex patterns for four different stages of filteration
regex_1 = re.compile("((?:[0-3]?\d)[\/|-](?:[01]?\d)[\/|-](?:(?:[1|2][0|9])?\d\d(?!\d))|(?:[01]?\d)[\/|-](?:[0-3]?\d)[\/|-](?:(?:[1|2][0|9])?\d\d(?!\d)))")
regex_2 = re.compile("((?:[0-3]?\d).{1,2}(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec|January|February|March|April|May|June|July|August|September|October|November|December).{1,2}(?:(?:[1|2][0|9])?\d\d(?!\d)))|((?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec|January|February|March|April|May|June|July|August|September|October|November|December).{1,2}(?:[0-3]?\d).{1,2}(?:(?:[1|2][0|9])?\d\d(?!\d)))")
regex_3 = re.compile("(?:\d{1}|\d{2})[\/]\d{4}")
regex_4 = re.compile("(?<!\/)\d{4}")

# =========================================================================================================================
# step one filteration process
# =========================================================================================================================

empty_line_numbers_list, result_dict = filter_data(regex_1, total_line_numbers_list, empty_line_numbers_list, result_dict)

# =========================================================================================================================
# step two filteration process
# =========================================================================================================================

total_line_numbers_list.clear()
total_line_numbers_list = [line for line in empty_line_numbers_list]
empty_line_numbers_list.clear()

empty_line_numbers_list, result_dict = filter_data(regex_2, total_line_numbers_list, empty_line_numbers_list, result_dict)

# =========================================================================================================================
# step three filteration process
# =========================================================================================================================

total_line_numbers_list.clear()
total_line_numbers_list = [line for line in empty_line_numbers_list]
empty_line_numbers_list.clear()

empty_line_numbers_list, result_dict = filter_data(regex_3, total_line_numbers_list, empty_line_numbers_list, result_dict)

# =========================================================================================================================
# step four filteration process
# =========================================================================================================================

total_line_numbers_list.clear()
total_line_numbers_list = [line for line in empty_line_numbers_list]
empty_line_numbers_list.clear()

empty_line_numbers_list, result_dict = filter_data(regex_4, total_line_numbers_list, empty_line_numbers_list, result_dict)
    
# =========================================================================================================================
# date data filteration completed
# =========================================================================================================================

# dictionary to store the normalized date and corresponding line number
normalized_date_dict = dict()

# date Normalization
for key,values in result_dict.items():
    # setting default month to January if month is missing
    # setting default day to 1 if day is missing
    dates = parser.parse(values, default = datetime(2022, 1, 1))

    # if the parser takes the year in two digit like 71 and by mistakeconvert that into 2071 instead of 1971 then this logic will take care of the conversion
    if dates.year > datetime.now().year:
        temp_values = f"{values[:-2]}19{values[len(values)-2:]}"
        dates = parser.parse(temp_values, default = datetime(2022, 1, 1))

    # in normalized date dictionary we are making the date as key so that it will be sorted in ascending order by default and also the corresponding document line numbers as well
    normalized_date_dict[dates] = key

chronological_date_list = normalized_date_dict.keys()
chronological_date_list = sorted(chronological_date_list)

# =========================================================================================================================
# Normalizing and sorting of line number based on chronological ascending order of the date is completed
# =========================================================================================================================

# creating new document with arranged lines/ events based on chronological order of the date

f = open("Chronological_Medical_Data_OUTPUT.txt", "w")
f.write(
'''
######################################################################################################
Events arranged in chronological order along with the corresponding line number from "dates.txt" file
######################################################################################################

'''
    )

for date in chronological_date_list:
    line_data = linecache.getline("./Raw_Medical_Data_INPUT.txt", normalized_date_dict[date])
    f.write(
f'''
Line-{normalized_date_dict[date]}
======================================================================================================
'''
    )
    f.write(f"{line_data}\n")

f.close()
```
